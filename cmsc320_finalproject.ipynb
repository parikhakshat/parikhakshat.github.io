## **Introduction**

## **Exploratory Data Analysis**

### **Statistical Method 1 (Z-Test for Outliers)**
From our basic data exploration and summary statistics, we can see that the number of samples for Acute Myeloid Leukemia is significantly larger than the population mean number of samples for leukemia types which is 12.8. Thus, we can use a Z-Test to determine whether the number of samples for Acute Myeloid Leukemia is an outlier in the dataset.

Z-Test Hypotheses

H0: The number of samples for Acute Myeloid Leukemia is not an outlier.

HA: The number of samples for Acute Myeloid Leukemia is an outlier.

Our significance level for this test will be 0.05.

Run the following Python code to perform the Z-Test on our dataset by first calculating the Z-Score for the number of samples for the AML type and then comparing whether the Z-Score is greater than some critical value.

```python
#Statistical Method 1: Z-Test to Determine Outliers
z_score = ((len(df[df['type'] == 'AML']) - type_counts.mean()) / np.std(type_counts))
print(f"Z-Score: {z_score}")
critical_value = stats.norm.ppf(1-0.05/2)
if z_score >  critical_value:
    print("Reject Null Hypothesis")
else:
  print("Fail to Reject Null Hypothesis")
```

```python
#Bar chart to show distribution of Leukemia types and possible outliers
figure = type_counts.plot(kind='barh')
figure.set_title("Number of Samples versus Type of Leukemia")
figure.set_xlabel("Number of Samples")
figure.set_ylabel("Type of Leukemia")
plt.show()
```

Using the Z-Test, we can determine that there was statistically signficant evidence to reject the null hypothesis and conclude that the number of samples for Acute Myeloid Leukemia is indeed an outlier in this dataset. Our conclusion is also further supported in the figure displayed above, where it shows that all the other leukemia types have about 10 samples, but the Acute Myeloid Leukemia type has more than double the number of samples. Finally, our conclusion is supported by current medical research, where the Mayo Clinic reported that by far the most common type of leukemia is Acute Myeloid Leukemia (https://www.mayoclinic.org/diseases-conditions/leukemia/symptoms-causes/syc-20374373).



### **Statistical Method 2 (ANOVA Test)**
From our summary statistics, we can see that there are over 22283 gene probes (features) in our dataset, and we aim to use the gene expression levels from these features to train a machine-learning model that can classify different types of leukemia. We would like to determine if there are major differences in these gene expression levels between all the different types of leukemia as that might provide information on what features are associated with each type of leukemia. Thus, we can perform an ANOVA test among the different types of leukemia and the gene expression levels to determine if there is a statistical difference between the different types.

Our code below can be changed to work with any gene in the dataset. For this ANOVA test, we will be focusing primarily on the RUNX1 gene as the American Cancer Society reported that it was the most common gene among patients (https://www.cancer.org/cancer/types/acute-myeloid-leukemia/causes-risks-prevention/what-causes.html).

ANOVA Test Hypotheses

H0: There is no significant correlation between the type of Leukemia and the gene expression levels for the RUNX1 gene.

HA: There is a significant correlation between the type of Leukemia and the gene expression levels for the RUNX1 gene.

Our significance level for this test will be 0.05.

Run the following Python code to perform an ANOVA test by first getting the gene expression levels for the desired gene for each type of leukemia. Then, the code just performs an ANOVA test and checks whether the P-Value is below our significance level.

```python
#Statistical Method 2: ANOVA Test
gene_name = "210365_at" #Converted RUNX1 to probe name

#Extract gene expression levels for each type of leukemia
bone_marrow_cd34_col = df[df['type'] == "Bone_Marrow_CD34"][gene_name]
pbsc_cd34_col = df[df['type'] == "PBSC_CD34"][gene_name]
bone_marrow_col = df[df['type'] == "Bone_Marrow"][gene_name]
pb_col = df[df['type'] == "PB"][gene_name]
aml_col = df[df['type'] == "AML"][gene_name]

#Calculate P-Value using an ANOVA test
f_statistic, p_value = stats.f_oneway(bone_marrow_cd34_col, pbsc_cd34_col, bone_marrow_col, pb_col, aml_col)

print(f"P-Value: {p_value}")

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
  print("Fail to Reject Null Hypothesis")
```

```python
# Box and whisker plot of all the different gene expression levels for RUNX1
plt.figure(figsize=(10, 10))

plt.boxplot([bone_marrow_cd34_col, pbsc_cd34_col, bone_marrow_col, pb_col, aml_col],
             labels=['Acute Leukemia CD34', 'PB Stem Cell',
                     'Bone Marrow', 'Peripheral Blood', 'AML'])

plt.title(f'Gene Expression Levels for RUNX1 Gene')
plt.xlabel('Leukemia Type')
plt.ylabel('Expression Level (RPM)')
plt.show()
```

Using the ANOVA test, we can determine that there was statistically signficant evidence to reject the null hypothesis and conclude that there is a correlation between the type of leukemia a person has and the gene expression level of the RUNX1 gene. This is further supported in our box and whisker plot, where we can see that each type of leukemia has a different median gene expression level (orange line on plot). This analysis shows us that the gene expression levels do vary across the types of leukemia, meaning that these features do have some pattern distinct to each type that can be determined using machine-learning models.

### **Statistical Method 3: Two Sample T-Test**

Another relationship we can investigate is if any genes themselves have correlated gene expression levels. This is important for us because we can use this info to select more features from our dataset to include when training our final model.

For this T-Test, we will be focusing primarily on if there is a difference between the gene expression levels of the RARA and DDR1 genes as the American Cancer Society reported that these genes were common among leukemia patients (https://www.cancer.org/cancer/types/acute-myeloid-leukemia/causes-risks-prevention/what-causes.html). Our code can be modified to use the T-Test for any two genes in our dataset.

Two-Sample T-Test Hypotheses

H0: There is a significant difference among the gene expression levels between the RARA and DDR1 genes.

HA: There isn't a significant difference among the gene expression levels between the RARA and DDR1 genes.

Our significance level for this test will be 0.05.


Run the following Python code to extract the gene expression levels for these two genes and then perform a T-Test on the data. The code then uses the resulting P-Value and compares it to our significance level.


```python
#Statistical Method 3: Two Sample T-Test
stat, p_value = stats.ttest_ind(df['209458_x_at'], df['211745_x_at']) #Convert the gene names to probe names
print(f"P-Value: {p_value}")

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
  print("Fail to Reject Null Hypothesis")
```

```python
#Correlation Matrix between RARA and DDR1 and some other genes
cols = ["209458_x_at", "211745_x_at", "210365_at", '1007_s_at', '1053_at', '117_at', '121_at', '1255_g_at',
       '1294_at', '1316_at', '1320_at', '1405_i_at', '1431_at', '1438_at']

#Calculate correlation matrix
corr_matrix = df[cols].corr()

#Print heatmap of correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', cbar=True, fmt=".2f",
            cbar_kws={'shrink': 0.8, 'label': 'Correlation Coefficient'})
plt.title(f"Correlation Matrix between 14 Genes (RARA, DDR1, etc)")
plt.show()
```

Using the two-sample T-Test, we can determine that there isn't statistically signficant evidence to reject the null hypothesis as the p-value was larger than the significance level. This means we can conclude that there is some correlation or difference between the gene expression levels of the RARA and DDR1 genes.

This is further supported in our graph, which is the correlation matrix of some of the genes in our dataset. We could not produce a full correlation matrix, effectively, as our dataset includes over 22000 genes (features), leading to significant computing time. The correlation matrix shows that between RARA (209458_x_at) and DDR1 (211745_x_at), the pearson correlation coefficient is one (actual value: 0.997), indicating a strong positive relationship between these two genes. As our graph and t-test show, we are able to conclude that there is some difference between the gene expression levels of the RARA and DDR1 genes.

**Random Forest/Decision Tree Classifiers**

The first classifier models we will investigate are Random Forest/Decision Tree classifiers. A Decision Tree classifier uses the examples from the training data to create a decision tree that maps the possible classes through a series of choices of the features. Random Forest is just an ensemble method that uses multiple decision trees to come to a singular result.

The following code trains these classifiers on our training data and evaluates their performance on the testing data.

```python
#Import packages needed for primary analysis
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import r2_score
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
```

The following Python code extracts the features and output data from our dataset, and then the code splits the features and output into testing and training portions.

```python
#Set up random state for the ML models
random_state = 42

#X is the gene expression levels from our dataset which is every column except the first and last two
X = df.iloc[:, 1:-2]

#Y is the target label which in our dataset is the type of leukemia
y = df['mapped_type']

#Split the dataset into testing and training portions
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)
```

We use this code to scale the gene expression levels for each gene probe (feature) in the dataset.

```python
#Use standard scaler to scale all the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

The following code creates the Random Forest/Decision Tree classifiers and trains the models on our training data. For Random Forest, we have set n_estimators to 2 as during our testing, we performed an iterative search over this hyperparameter and determined this to be the best value.

```python
#Create and fit Random Forest classifier on training data
rf_classifier = RandomForestClassifier(n_estimators = 2, criterion = 'entropy', random_state = random_state)
rf_classifier.fit(X_train, y_train)

#Create and fit DecisionTree classifier on training data
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)
```

We use this code to get predictions for the Random Forest model and calculate the f1-score, accuracy, and r^2 score. These will be the main metrics we will use to evaluate the performance of all our machine-learning models on the testing data.

```python
#Get the label predictions (type of leukemia) for all samples in testing dataset
rf_pred = rf_classifier.predict(X_test)

#Calculate the accuracy, f1-score, and r^2 of the Random Forest model on the testing dataset
rf_accuracy = accuracy_score(y_test, rf_pred)
rf_f1 = f1_score(y_test, rf_pred, average='weighted')
rf_r2 = r2_score(y_test, rf_pred)

print(f"Accuracy (Random Forest): {rf_accuracy}")
print(f"F1-Score (Random Forest): {rf_f1}")
print(f"Coefficient of Determination (Random Forest): {rf_r2}")
```

We use this code to get predictions for the Decision Tree model and calculate the f1-score, accuracy, and r^2 score.

```python
#Get the label predictions (type of leukemia) for all samples in testing dataset
dt_pred = dt_classifier.predict(X_test)

#Calculate the accuracy, f1-score, and r^2 of the Decision Tree model on the testing dataset
dt_accuracy = accuracy_score(y_test, dt_pred)
dt_f1 = f1_score(y_test, dt_pred, average='weighted')
dt_r2 = r2_score(y_test, dt_pred)

print(f"Accuracy (Decision Tree): {dt_accuracy}")
print(f"F1-Score (Decision Tree): {dt_f1}")
print(f"Coefficient of Determination (Decision Tree): {dt_r2}")
```

**Support Vector Machine Classifier**

The next classifier model we will use is the Support Vector Machine classifier. The Support Vector Machine classifier works by finding a hyperplane that has the maximal distance between two classes. The Support Vector Machine classifier can only be trained on binary classification tasks, so we need to use the OneVsRestClassifier on top of SVM to turn our multi-class classification task into a binary classification task. This is done by creating separate binary classifiers for each one of the classes in the dataset. We also had to use the SVM kernel trick which allows us to use higher-dimensional input with Support Vector Machines. The kernels we chose to test are rbf and polynomial. For the SVM model with the RBF kernel, I have set the gamma to 0.5 and C to 1 as during testing, we performed an iterative search over these hyperparameters and found these to be the best values. For the SVM model with the poly kernel, I have set the degree to 2 and C to 1 as during testing, we performed an iterative search over these hyperparameters and found these to be the best values.

Run the following code to create the two SVM classifiers, train them on our training data, and evaluate their performance on the testing data.

```python
#Create OneVsRestClassifier instance to do multi-class classification with SVM and rbf kernel
rbf = OneVsRestClassifier(svm.SVC(kernel='rbf', gamma=0.5, C=1))
#Fit the model on the training data
rbf.fit(X_train, y_train)

#Create OneVsRestClassifier instance to do multi-class classification with SVM and poly kernel
poly = OneVsRestClassifier(svm.SVC(kernel='poly', degree=2, C=1))
#Fit the model on the training data
poly.fit(X_train, y_train)
```

```python
#Get predictions from the SVM poly and rbf models on the testing data
poly_pred = poly.predict(X_test)
rbf_pred = rbf.predict(X_test)
```

```python
#Calculate the accuracy, f1-score, and r^2 of the SVM (Polynomial Kernel) model on the testing dataset
poly_accuracy = accuracy_score(y_test, poly_pred)
poly_f1 = f1_score(y_test, poly_pred, average='weighted')
poly_r2 = r2_score(y_test, poly_pred)

print(f'Accuracy (SVM Polynomial Kernel): {poly_accuracy}')
print(f'F1-Score (SVM Polynomial Kernel): {poly_f1}')
print(f"Coefficient of Determination (SVM Polynomial Kernel): {dt_r2}")

print("\n")

#Calculate the accuracy, f1-score, and r^2 of the SVM (RBF Kernel) model on the testing dataset
rbf_accuracy = accuracy_score(y_test, rbf_pred)
rbf_f1 = f1_score(y_test, rbf_pred, average='weighted')
rbf_r2 = r2_score(y_test, rbf_pred)

print(f'Accuracy (SVM RBF Kernel): {rbf_accuracy}')
print(f'F1-Score (SVM RBF Kernel): {rbf_f1}')
print(f"Coefficient of Determination (SVM RBF Kernel): {rf_r2}")
```

Finally, we have set the value for K to be 18 in this model because during our testing, we performed an iterative search over this hyperparameter and found this to be the best value.

Run the following code to create the KNN classifier, train the model on our dataset, and evaluate its performance on the testing data.

```python
#Create OneVsRestClassifier instance to do multi-class classification with KNN
knn = OneVsRestClassifier(KNeighborsClassifier(18))
#Fit the model on training data
knn.fit(X_train, y_train)

#Get predictions of model on testing data
knn_pred = knn.predict(X_test)

#Calculate the accuracy, f1-score, and r^2 of the KNN model on the testing dataset
knn_accuracy = accuracy_score(y_test, knn_pred)
knn_f1 = f1_score(y_test, knn_pred, average='weighted')
knn_r2 = r2_score(y_test, knn_pred)

print(f"Accuracy (KNN): {knn_accuracy}")
print(f"F1 (KNN): {knn_f1}")
print(f"Coefficient of Determination (KNN): {knn_r2}")
```

The Multilayer Perceptron classifier creates a fully-connected deep neural network that can be used for multi-class classification tasks. Run the following code to create the MLP model, train the classifier on our dataset, and evaluate its performance on our training data.

## **Insights and Conclusions**
In our primary analysis, we trained several different types of multi-class classifiers to use the gene expression levels of a patient to predict the type of leukemia they have. We then evaluated the classifiers on a held-out testing dataset, determining the accuracy, F1-score, and R^2 score for all the models. Using these metrics, we can determine which type of classifier had the strongest performance, which was the Multilayer Perceptron classifier. This model achieved the highest accuracy of 0.92, the highest F1-score of 0.92, and the highest coefficient of determination of 0.96. A R^2 score of 0.96 means that approximately 96% of the variation in the dependent variable (type of leukemia) was predicted by the machine-learning model. The second-best model is the Random Forest classifier as it achieved the second highest values for accuracy, F1-score, and R^2 score at 0.85, 0.84, and 0.65 respectively. The K-Nearest Neighbor, SVM, and Decision Tree classifiers significantly underperformed when compared to the other classifiers. The K-Nearest Neighbor and Decision Tree classifiers achieved low accuracies and F1-scores around 0.6. With these metrics and the results from the confusion matrix visualization, it is evident that from our primary analysis, the best classification model for our dataset is the Multilayer Perceptron classifier.

The Multilayer Perceptron classifier performed the best for our project because our dataset involved over 22000 features, which meant that the dataset was very high-dimensional. With high-dimensional datasets, there can be a phenomena named “curse of dimensionality”, which is when models underperform because the high-dimensional data is too sparse. The data sparsity negatively affects K-Nearest Neighbors because in high-dimensional spaces, the concept of “distances” between points becomes less meaningful. Decision Tree classifiers tend to overfit on high-dimensional data too as the tree becomes substantially large and tends to memorize the noise in the dataset rather than capture the underlying pattern. In our dataset, the relationships between gene expression levels and the type of leukemia might also be nonlinear as evidenced in our exploratory data analysis. This means that classifiers like Support Vector Machines would perform poorly on our dataset as they would need to use some kernel trick to model these nonlinear relationships. Multilayer Perceptron classifiers avoid all the other pitfalls that these other models fall into. This is because Multilayer Perceptrons are a neural network made up of several connected layers of neurons, which can handle the high-dimensional data from our dataset. Between each layer in the network, there is also an activation function which allows the network to model nonlinear relationships. Thus, since Multilayer Perceptron classifiers perform well on multi-class classification with high-dimensional data and nonlinear relationships, they are the best model we should use to predict the type of leukemia a patient has.

With our Multilayer Perceptron model, we can get a 92% accuracy on the testing portion of our dataset, which means that in most cases, our model can accurately predict the type of leukemia a patient has based on their gene expression levels. We can connect this back to our research question and now determine that there is a relationship between the gene expression levels of a human and different types of leukemia. Our machine-learning model is able to discern these patterns from our dataset and accurately predict the type of leukemia a patient has. This is highly important in the medical field because our machine-learning model can be used to do quick and early diagnosis for leukemia. As mentioned before, many traditional diagnostic tests for leukemia only work after certain symptoms appear, which can be significantly far into a patient’s treatment. Our machine-learning model just needs gene expression levels which can be measured for a human at any time, allowing us to use our machine-learning model to screen for leukemia well ahead of other diagnostic tests. Moreover, our machine-learning model is also much faster than other leukemia diagnostic tests which can take anywhere from a week to a month on average to receive results from. On the other hand, our model just needs the gene expression levels, and then, the actual testing part of the levels can happen with the model in a few seconds. Finally, an idea to improve our project is to see whether we can achieve a similar level of accuracy by using fewer features as this would make it much easier to train and use our models. We can also try adapting our approach to other cancer datasets and see if we are able to discern similar relationships in the data. In the end, we were able to train a classifier model on our dataset that achieved a 92% accuracy in predicting the type of leukemia from the gene expression levels of a patient.
